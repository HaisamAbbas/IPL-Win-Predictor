### grouping the 1st innings, 2nd innings score in a particular match id
### lets say match id = 1, so inning 1 score = 207, inning 2 score = 172, like this:

```python
# grouping the 1st innings, 2nd innings score in a particular match id
# lets say match id = 1, so inning 1 score = 207, inning 2 score = 172, like this:

totalrun_df = deliveries.groupby(['match_id','inning']).sum()['total_runs'].reset_index()

totalrun_df
```

### Capturing only the first innings, as we will be predicting for the second innings 

```python
#### Capturing only the first innings, as we will be predicting for the second innings

totalrun_df = totalrun_df[totalrun_df['inning']==1]
totalrun_df['total_runs'] = totalrun_df['total_runs'].apply(lambda x:x+1)
totalrun_df
```

### Matches column

```python
# Matches column
matches.columns
```

### Merging the total first innigs score df with the matches df, where left side merging is done on "id" column of the matches and right side merging is done on "match_id" column of the totalrun_df

```python
'''
Merging the total first innigs score df with the matches df,
where left side merging is done on "id" column of the matches
and right side merging is done on "match_id" column of the totalrun_df

'''

match_df = matches.merge(totalrun_df[['match_id','total_runs']],
                        left_on='id',right_on='match_id')

match_df
```

```python
match_df['team1'].unique()
```

```python
teams = [
    'Kolkata Knight Riders',
    'Chennai Super Kings', 
    'Rajasthan Royals',
    'Mumbai Indians',
    'Deccan Chargers'
]
```

### Replacing the Delhi Daredevils/Deccan Chargers with Delhi Capitals

```python
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hydrabad')

match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hydrabad')

match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
match_df['team1'].unique()
```

```python
# will consider only frequently occuring teams,
# which are mentioned in the team list

match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]

match_df['team1'].unique()
```

```python
match_df.shape
```

```python
match_df.head()
```

### Checking the matches which resulted in dl method

### Ignoring the rows which were DL method

```python
### Ignoring the rows which were DL method

match_df = match_df[match_df['dl_applied']==0]

## Considring the match_id,city,winner and total runs

match_df = match_df[['match_id','city','winner','total_runs']]

match_df
```

```python
deliveries.head()
```

### merging matchdf with deliveries on match_id

```python
delivery_df = match_df.merge(deliveries,on='match_id')

delivery_df.head()
```

```python
delivery_df.columns
```

```python
# Considering the second innings

delivery_df = delivery_df[delivery_df['inning'] ==2]
delivery_df.head()
```

```python
delivery_df.shape
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']

delivery_df.head()                                               
```

```python
# runs left

delivery_df['runs_left'] = delivery_df['total_runs_x']-delivery_df['current_score']

delivery_df.head()
```

```python
delivery_df['balls left'] = 126-(delivery_df['over']*6+delivery_df['ball'])

delivery_df
```

```python
list(delivery_df['player_dismissed'].unique())
```

### Filling with 0

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna('0')

# now we will convert this player_dismissed column into a boolean column
# if the player is not dismissed then it's 0 else its 1

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x=='0' else 1)

# converting into into string

delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')

delivery_df['player_dismissed'].unique()
```

```python
# Wickets fallen

wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values

delivery_df['wickets'] = 10-wickets

delivery_df
```

```python
# Current run rate

delivery_df['cur_run_rate'] = (delivery_df['current_score']*6)/(120-delivery_df['balls left'])

# required run rate

delivery_df['req_run_rate'] = (delivery_df['runs_left']*6)/(delivery_df['balls left'])

delivery_df.head()
```

```python
def resultfun(row):
    
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(resultfun,axis=1)
delivery_df.head()
```

```python
sn.countplot(delivery_df['result'])
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left',
                         'balls left','wickets','total_runs_x','cur_run_rate',
                        'req_run_rate','result']]

final_df.head()
```

```python
final_df.shape
```

```python
final_df.isnull().sum()
```

```python
# dropping of null values

final_df = final_df.dropna()

final_df.isnull().sum()
```

```python
final_df = final_df[final_df['balls left']!= 0]
```

```python
data = final_df.copy()

test = data['result']

train = data.drop(['result'],axis = 1)

train.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(train,test,test_size=0.2,random_state=1)

X_train.shape,X_test.shape
```

```python
X_train.columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn import metrics
```

```python
# Batting team, bowling team and city are categorical columns
# they will be converted to numeric using onehot encoder

cf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
# creating the Pipeline

pipe = Pipeline(steps=[
    ('step1',cf),
    ('step2',LogisticRegression(solver='liblinear'))
])

# fitting the training data

pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
print(metrics.accuracy_score(y_test,y_pred))
```

```python
pipe.predict_proba(X_test)[10]
```

```python
pipe2 = Pipeline(steps=[
    ('step1',cf),
    ('step2',RandomForestClassifier())
])

# fitting the training data

pipe2.fit(X_train,y_train)
```

```python
print(metrics.accuracy_score(y_test,pipe2.predict(X_test)))
```

```python
pipe2.predict_proba(X_test)[10]
```
